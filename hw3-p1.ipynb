```python
index=index.T
index
```

```python
gene.columns = index.iloc[0,].str.strip()
```

```python
label = pd.read_csv('Data/label.txt', sep=" ",header=None)
```

```python
label.columns = ['label_sign']
# replace positive with 1 and negative with 0
label['label_sign'] = label['label_sign'].apply(lambda x : 1 if x > 0 else 0)
label
```

```python
from sklearn.feature_selection import SelectKBest, f_classif , chi2
```

```python
selector = SelectKBest(score_func=chi2, k=60)
new_data = selector.fit_transform(gene, label)
mask = selector.get_support()
new_features = gene.columns[mask]
new_features
```

### (a) Simply describe your feature selection method.

Univariate Feature Selection
The simplest and fastest methods are based on univariate statistical tests. For each feature, measure how strongly the target depends on the feature using a statistical test like  𝜒2.

From the scikit-learn feature selection module, feature_selection.SelectKBest returns the K best features given some scoring function. For our classification problem, the module provides scoring functions:  𝜒2.

With SelectKBest, we define the number of features to keep, based on the score from the scoring function. Using .fit_transform(features, target) we get back an array with only the selected features.



在特徵選擇中，最適合在探索階段使用的就是有效的單變量選擇法。

第一題中 simple feature selection using naive one-by-one selection 我使用的是 sklearn中的單變量選擇法，
單變量選擇法的主要思路是根據某些統計檢驗的方法分別對每個變量進行檢驗，得到一組分數、p-value數據，然後我們排序選擇分數最高(或p-value最小等)的那些特徵。在sklearn中，分別針對不同的統計檢驗方法和不同的排序選擇標準提供了不同的工具，比如用於回歸問題的f_regression、mutual_info_regression分數，用於分類問題的f_classif、chi2、mutual_info_classf分數；以及用於特徵排序和選擇的SelectKBest、SelectPercentile、SelectFpr等。把這兩類工具相結合，就可以完成特徵選擇的任務了。

顧名思義，SelectKBest方法就是從所有特徵中挑選出最好的K個特徵組成新的特徵集。如何來定義最好呢？這要取決於我們傳入的評分函數，該方法默認會使用f_classif，這種方法適用於為分類模型評估特徵，與之對應的是f_regression，它適用於回歸模型的特徵評估。
但在這題我使用的是Chi-squared test，然後將分數從高到低排序，選出分數最高的前60個Feature，作為我Feature Selection所挑選出來的Feature結果

chi2:

Chi-squared stats of non-negative features for classification tasks.
